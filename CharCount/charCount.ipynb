## Number of words:

```python
Wrds = re.findall(r'[\w]+', ST1)
```

```python
len(Wrds)
```

```python
for idx,val in enumerate(Wrds):
    print(idx,':',val)
```

```python
Wrds[1]
```

```python
UniqWrds = set(Wrds)
```

```python
UniqWrds
```

## Number of occurances of each word

```python
UniqWrdsDict = {i : 0 for i in UniqWrds}
```

```python
UniqWrdsDict
```

```python
for i in UniqWrds:
    for j in Wrds:
        if i==j:
            UniqWrdsDict[i] += 1
```

```python
UniqWrdsDict
```

```python
for k,l in UniqWrdsDict.items():
    print('Word :',k,'|','Number of occurence:',l)
```

## Special chars:

```python
spcChr = re.findall(r'[\W]', ST1)
```

```python
spcChr
```

```python
# #spcChr1 = spcChr.str.strip(' ')
# SPstr = ''.join(str(e) for e in spcChr)
# SPstr
SC = []
for i in spcChr:
    if (i != ' '):
        SC.append(i)
```

```python
SC2 = []
for i in SC:
    if (i != '\n'):
        SC2.append(i)
```
